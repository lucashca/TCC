```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sb

%matplotlib inline
```

```python
df = pd.read_csv('../datasets/datawine.csv')
print(df)
```

```python
#b.pairplot(df)
```

```python
X = np.array(df)
X
```

# K-Means

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5,random_state = 0,n_jobs=1,max_iter=10000)
```

```python
kmeans.fit(X)
```

```python
print(kmeans.labels_)
```

# Comparando as Classes Criadas

```python

df['K-classes'] = kmeans.labels_
```

```python
df
```

```python
sb.pairplot(df,hue='K-classes')
```
